# Linear Regression (Buildings)

#### Libraries
____

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder

import statsmodels.formula.api as sm

import pickle

%matplotlib inline
```

#### Load dataset
___

```python
dataset = '../../data/commercial_data_clear.csv'
columns = ['ptype', 'floor_area', 'build_year', 'occupy', 'star', 'site_eui', 'ngas', 'water_use']


df = pd.read_csv(dataset, sep=',', usecols=columns)

cond1 = df['ptype'] == 'K-12 School'

building = df[cond1]

building.head(5)
```

#### Define X, Y variables
___

```python
#x_columns = ['ptype', 'floor_area', 'build_year', 'occupy', 'star', 'ngas', 'water_use', 'total_fuel']
x_columns = ['floor_area', 'build_year', 'occupy', 'star', 'ngas', 'water_use']
y_columns = ['site_eui']

X = building[x_columns].values
Y = building[y_columns].values
```

#### Label Encoding
___

```python
building['ptype'].unique()
```

```python
building['ptype'].nunique()
```

```python
plt.figure(figsize=(16,6))
ax = sns.countplot(x='ptype', data=building, palette='YlGnBu')
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.tight_layout()

plt.show()
```

```python
label_encoder_ptype = LabelEncoder()
X[:, 0] = label_encoder_ptype.fit_transform(building['ptype'])


label_hotencoder_ptype = OneHotEncoder(categorical_features = [0])
X = label_hotencoder_ptype.fit_transform(X).toarray()

print(X)
```

### Avoiding the Dummy Variable Trap
____


```python
X = X[:, 1:]
```

#### Split dataset (TRAIN/TEST) data
____

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 0)
```

```python
regressor = LinearRegression() 
regressor.fit(X_train, Y_train)
```

#### Save model into disk
_____

```python
pickle.dump(regressor, open('buildings.pkl','wb'))
```

#### Fitting Simple Linear Regression to the Training Set
_____

```python
y_prediction = regressor.predict(X_test)
```

#### Real x Predict
_____

```python
pd.DataFrame({
    'Real' : Y_test[:,0],
    'Predict' : y_prediction[:,0]

})
```

#### Model Score
__________

```python
print(f"Training Data Score: {regressor.score(X_train, Y_train)}")
print(f"Testing Data Score: {regressor.score(X_test, Y_test)}")
```

### Backward Elimination

```python
X = np.append(arr = np.ones((len(X), 1)).astype(int), values = X, axis =1)
```
