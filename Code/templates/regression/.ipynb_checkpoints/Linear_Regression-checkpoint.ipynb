#### Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder

import statsmodels.formula.api as sm


%matplotlib inline
```

#### Load dataset
___

```python
dataset = '../../../Building Performance Database/data/building.txt'
columns = ['ptype', 'floor_area', 'floor_report', 'year', 'weather_site', 'weather_source', 'eui_site', 'direct_ghg', 'indirect_ghg', 'total_ghg']

building = pd.read_csv(dataset, sep='\t', usecols=columns)

building.head(5)
```

#### Data cleanup
___

```python
plt.figure(figsize=(16,6))

sns.heatmap(building.isnull(), yticklabels=False, cbar=False, cmap='viridis')
plt.show()
```

```python
building.dropna(axis='columns', how='all', inplace=True)
building.dropna(inplace=True)
building.head()
```

#### Define X, Y variables
___

```python
x_columns = ['ptype', 'floor_area', 'floor_report', 'year', 'weather_site', 'weather_source', 'direct_ghg', 'indirect_ghg', 'total_ghg']
y_columns = ['eui_site']

X = building[x_columns].values
Y = building[y_columns].values
```

#### Label Encoding
___

```python
building['ptype'].unique()
```

```python
label_encoder_ptype = LabelEncoder()
X[:, 0] = label_encoder_ptype.fit_transform(building['ptype'])


label_hotencoder_ptype = OneHotEncoder(categorical_features = [0])
X = label_hotencoder_ptype.fit_transform(X).toarray()

print(X)
```

### Avoiding the Dummy Variable Trap
If youre interested in checking out the documentation, you can find it here. Now, as we already discussed, depending on the data we have, we might run into situations where, after label encoding, we might confuse our model into thinking that a column has data with some kind of order or hierarchy, when we clearly dont have it. To avoid this, we OneHotEncode that column
____


```python
X = X[:, 1:]
```

#### Split dataset (TRAIN/TEST) data
____

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 0)
```

```python
regressor = LinearRegression() 
regressor.fit(X_train, Y_train)
```

#### Fitting Simple Linear Regression to the Training Set
_____

```python
y_prediction = regressor.predict(X_test)
```

#### Real x Predict
_____

```python
pd.DataFrame({
    'Real' : Y_test[:,0],
    'Predict' : y_prediction[:,0]

})
```

#### Model Score
__________

```python
print(f"Training Data Score: {regressor.score(X_train, Y_train)}")
print(f"Testing Data Score: {regressor.score(X_test, Y_test)}")
```

### Backward Elimination
___

```python
X = np.append(arr = np.ones((len(X), 1)).astype(int), values = X, axis =1)
```

```python
X_optimized = X[:, [-1, -2]]

significante_level = 0.05

regressor_OLS = sm.OLS(endog = Y, exog = X_optimized).fit()
```

```python
regressor_OLS.summary()
```
