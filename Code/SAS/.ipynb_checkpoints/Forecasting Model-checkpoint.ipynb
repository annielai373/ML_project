# General Forecasting Models
______

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
emission = pd.read_csv('../../Greenhouse/data/sample_emission_data.csv', index_col='Year', parse_dates=True)
emission.index.freq = 'AS'
```

```python
emission.head()
```

```python
emission.info()
```

### Plot dataset
_______

```python
emission.plot(figsize=(12,6));
```

#### Decompose Time Series Data
____

```python
from statsmodels.tsa.seasonal import seasonal_decompose

result = seasonal_decompose(emission['Emission'],model='multiplicative')

from pylab import rcParams
rcParams['figure.figsize'] = 12,5

result.plot();
```

### Split dataset
##### Test dataset lenght should be the same as prediction
_______

```python
train_data = emission[:24]    # or by date .loc[:'1990-01-01']
test_data = emission[23:]
```

#### Create Model
___

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing

fitted_model = ExponentialSmoothing(train_data['Emission'], 
                                    trend='mul').fit()
```

### Make predictions using Test Dataset
________

```python
test_predictions = fitted_model.forecast(5) # 5 years
test_predictions
```

```python
train_data['Emission'].plot(legend=True, label='TRAIN', figsize=(12,6));
test_data['Emission'].plot(legend=True, label='TEST', figsize=(12,6));
```

### Evaluate prediction performance using Test data
________

```python
train_data['Emission'].plot(legend=True, label='TRAIN', figsize=(12,6));
test_data['Emission'].plot(legend=True, label='TEST', figsize=(12,6));
test_predictions.plot(legend=True, label='PREDICTION');
```

### Model Evaluation
________

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

np.sqrt(mean_squared_error(test_data, test_predictions))
```

```python
mean_absolute_error(test_data, test_predictions)
```

```python
# lets compare errors with the average value of the test dataset
test_data.describe()
```

### Forecast into the future
________

```python
final_model = ExponentialSmoothing(emission['Emission'], 
                                    trend='mul').fit()
```

```python
forecast_predictions = final_model.forecast(5)
```

```python
emission['Emission'].plot(legend=True, label='REAL', figsize=(12,6));
forecast_predictions.plot(legend=True, label='PREDICTION');
```
